```python
import numpy as np
import re
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import accuracy_score, classification_report 
from sklearn.linear_model import LogisticRegression
import re
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
data = []
with open("C:\\Users\\Sarthak Tyagi\\Downloads\\train_data.txt",'r',encoding = 'utf-8') as file:
    for line in file:
        parts = line.strip().split(' ::: ')
        if len(parts)==4:
            movie_id, title, genre, plot = parts
            data.append([movie_id,title,genre,plot])
```

```python
df = pd.DataFrame(data, columns=['id','title','genre','plot'])
df.head()
```

```python
def clean_text(text):
    text = re.sub(r'\W',' ',text)
    text = text.lower()
    text = re.sub(r'\s+',' ',text)
    return text
```

```python
df['cleaned_plot']=df['plot'].apply(clean_text)
```

```python
df
```

```python
df.drop('plot',axis='columns')
```

```python
X = df['cleaned_plot']
y = df['genre']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 64)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline 
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
